```python
import pandas as pd
```

```python
data = pd.read_csv("boston_housing_data2.csv", sep=",", header=None)
```

```python
data.rename(columns={1:"CRIM", 2:"ZN", 3:"INDUS", 4:"CHAS", 5:"NOX", 6:"RM", 7:"AGE", 8:"DIS", 9:"RAD", 10:"TAX", 11:"PTRATIO", 12:"B", 13:"LSTAT", 14:"MEDV"}, inplace = True )
```

```python
data.drop(0, axis=1, inplace = True)
```

```python
data
```

```python
median_home_value = data.MEDV.median()*1000
```

```python
"Median Home Value: ${:.2f}".format(median_home_value)
```

```python
average_home_value = data.MEDV.mean()*1000
"Average Home Value: ${:.2f}".format(average_home_value)
```

```python
import matplotlib.pyplot as plt
```

```python
x = data.CRIM.sort_values()
```

```python
y = data.PTRATIO.sort_values()
```

```python
markers_on = [data[data.CHAS == 1]][0]
```

```python
markers_on
```

```python
hly = markers_on.PTRATIO.sort_values()
hlx = markers_on.CRIM.sort_values()
```

```python
plt.title("per-capita crime rate and the pupil-teacher ratio")
plt.xlabel("Per Capita Crime Rate")
plt.ylabel("Pupil to Teacher Ratio")
plt.scatter(x, y, color='red')
plt.plot(x, y, color='blue')
plt.scatter(hlx, hly, color="green")
plt.show()
```

```python
x = data.B.sort_values()
y = data.DIS.sort_values()
```

```python
plt.title("proportion of black citizens and the distance to employment centers")
plt.xlabel("proportion of black citizens")
plt.ylabel("distance to employment centers")
plt.scatter(x, y, color='red')
plt.plot(x, y, color='blue')
plt.show()
```
