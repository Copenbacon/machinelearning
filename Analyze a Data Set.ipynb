## Comparing Crime Rate Per Capita against Pupil to Teacher Ratio
with markers highlighting suburbs near the Charles River

```python
x = data.CRIM.sort_values()
```

```python
y = data.PTRATIO.sort_values()
```

```python
markers_on = [data[data.CHAS == 1]][0]
```

```python
markers_on
```

```python
hly = markers_on.PTRATIO
hlx = markers_on.CRIM
```

```python
plt.title("per-capita crime rate and the pupil-teacher ratio")
plt.xlabel("Per Capita Crime Rate")
plt.ylabel("Pupil to Teacher Ratio")
plt.scatter(x, y, color='red')
plt.plot(x, y, color='blue')
plt.scatter(hlx, hly, color="green")
plt.show()
```

Crime rate spikes upward when pupil to teaher ratio exceeds about 25 or 30. Homes near the Charles River generally have less crime and more teachers per pupil, though some Charles River neighborhoods do have higher crime rates.

## Comparing Proportion of Black Citizens with Suburb Distance to Employment Centers

```python
x = data.B.sort_values()
y = data.DIS.sort_values()
```

```python
plt.title("proportion of black citizens and the distance to employment centers")
plt.xlabel("proportion of black citizens")
plt.ylabel("distance to employment centers")
plt.scatter(x, y, color='red')
plt.plot(x, y, color='blue')
plt.show()
```

Black citizens distance from employment centers spikes as the proportion of black citizens in a neighborhood grows over 37% (370/1000). Also, no neighborhood has more than 40% black citizens.

## Comparing Median Values of Homes sorted by NOx Distribution with Median Value of Homes sorted by Proportion of Non-Retail Businesses Acres Per Town

```python
y = data.sort_values(['NOX', 'MEDV'], ascending=[True, False]).MEDV
x = data.sort_values(['INDUS', 'MEDV'], ascending=[True, False]).MEDV
```

```python
plt.title("Median Value of Homes by NOx vs Median Value of Homes by Proportion of Non-Retail Businesses")
plt.ylabel("Median Value of Homes by NOx distribution")
plt.xlabel("Median Value of Homes by Proportion of Non-Retail Businesses")
plt.scatter(x, y, color='red')
plt.show()
```

Generally, homes produce a lower amount of NOx and are surrounded by more retail businesses.
